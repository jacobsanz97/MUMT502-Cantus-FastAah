```python
from midiutil.MidiFile import MIDIFile

def volpiano2midi(input_string):
    #create MIDI object
    mf = MIDIFile(1)    #1 track
    track = 0   # the only track

    time = 0    # start at the beginning
    mf.addTrackName(track, time, "Sample Track")
    mf.addTempo(track, time, 80)

    
    # add some notes
    channel = 0
    volume = 100
    
    #monk sound
    program = 53
    
    time = 0
    duration = 1
    mf.addProgramChange(track, channel, time, program)
    for character in input_string:    
        if character == "9":
            time = time + 1
            pitch = 55
            mf.addNote(track, channel, pitch, time, duration, volume)
        
        if character == "a":
            time = time + 1
            pitch = 57
            mf.addNote(track, channel, pitch, time, duration, volume)
            
        if character == "b":
            time = time + 1
            pitch = 59
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "c":
            time = time + 1
            pitch = 60
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "d":
            time = time + 1
            pitch = 62
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "e":
            time = time + 1
            pitch = 64
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "f":
            time = time + 1
            pitch = 65
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "g":
            time = time + 1
            pitch = 67
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "h":
            time = time + 1
            pitch = 69
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "j":
            time = time + 1
            pitch = 71
            mf.addNote(track, channel, pitch, time, duration, volume)
            
        if character == "k":
            time = time + 1
            pitch = 72
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "l":
            time = time + 1
            pitch = 74
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "m":
            time = time + 1
            pitch = 76
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "n":
            time = time + 1
            pitch = 77
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "o":
            time = time + 1
            pitch = 79
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "p":
            time = time + 1
            pitch = 81
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "q":
            time = time + 1
            pitch = 83
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "r":
            time = time + 1
            pitch = 55
            mf.addNote(track, channel, pitch, time, duration, volume)
   
        if character == "s":
            time = time + 1
            pitch = 84
            mf.addNote(track, channel, pitch, time, duration, volume)
   
    with open("output.mid", 'wb') as outf:
        mf.writeFile(outf)
    
input_string = "1---e--c--e---g--h---h--f--g---g--h--f---g---f7--e--d--e---g--f---d--e--f---e--d--c---d77-"
input_string2 = "9--9-9"
volpiano2midi(input_string)
```
